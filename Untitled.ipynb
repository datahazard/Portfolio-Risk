```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
MSFTPrices_csv = pd.read_csv('MSFTPrices.csv', parse_dates=['Date'])
```

```python
print(MSFTPrices_csv.head(5))
```

```python
MSFTPrices_csv = MSFTPrices_csv.sort_values(by='Date')
```

```python
print(MSFTPrices_csv.head(5))
```

```python
MSFTPrices_csv['Returns'] = MSFTPrices_csv['Adjusted'].pct_change()
```

```python
print(MSFTPrices_csv.head(5))
```

```python
MSFTPrices_csv['Returns'].plot()
plt.show()
```

```python
percent_return = MSFTPrices_csv['Returns']*100
returns_plot = percent_return.dropna()
```

```python
plt.hist(returns_plot, bins=75)
plt.show()
```
